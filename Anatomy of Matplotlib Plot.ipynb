### 0. Import Matplotlib and get it ready for Jupyter

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

### 1. Prepare Data

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
```

### 2. Setup Plot

```python
fig, ax = plt.subplots(figsize=(10, 10))
```

### 3. Plot Data

```python
ax.plot(x, y)
```

### 4. Customize Plot

```python
ax.set(
    title="simple_plot",
    xlabel="x-axis",
    ylabel="y-axis"
)
```

### 5. Save the plot

```python
fig.savefig("sample-plot.png")
```

### Making figures with Numpy arrays

```python

import numpy as np
```

```python
# create some data
x = np.linspace(0,10,100)
# start, stop, step
x
```

### Plot the data
Line Plot

```python
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

Scatter plot

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

another scatter plot

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

Bar Plot from Dictionary

```python
# creating data
fruit_prices = {
    "Apple":10,
    "Banana":8,
    "Orange":12
}
```

```python
fig, ax = plt.subplots()
ax.bar(fruit_prices.keys(), fruit_prices.values())
ax.set(title="Radium Fruit Shop", ylabel="Price $");
```

### Horizontal Bar Plot


```python
fig, ax = plt.subplots()
ax.barh(list(fruit_prices.keys()), list(fruit_prices.values()))
ax.set(title="Radium Fruit Shop", ylabel="Price $");
```

```python
# make some data
x = np.random.randn(1000)
x
```

### Histogram

```python
fig, ax = plt.subplots()
ax.hist(x);
```

### Multiple Plot
Subplot option 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(
                                            nrows = 2,
                                            ncols = 2,
                                            figsize = (10, 5)
)

# plot to each axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(fruit_prices.keys(), fruit_prices.values())
ax4.hist(np.random.randn(1000));
```

#### Subplot option 2 : index

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5)
                      )
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(fruit_prices.keys(), fruit_prices.values())
ax[1,1].hist(np.random.randn(1000));
```
